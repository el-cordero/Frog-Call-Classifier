```python
import librosa
import numpy as np
import glob
import os
import math
import pandas as pd
import re
import datetime
from datetime import datetime, timedelta
import soundfile as sf
```

```python
# ✅ Define folder containing audio files
AUDIO_DIR = os.getcwd().replace("/_scripts", "/_data/Audio/Full/Starlink_Group_7-13")
OUTPUT_DIR = os.getcwd().replace("/_scripts", "/_data/TrainingClips")
os.makedirs(OUTPUT_DIR, exist_ok=True)  # Create if doesn't exist
```

```python
spectrograms = os.getcwd().replace("/_scripts","/_results/Model/output/filestoextract_positives.csv")
spectrograms = pd.read_csv(spectrograms)
spectrograms.head()
spectrograms = os.getcwd().replace("/_scripts","/_results/Model/output/filestoextract_negatives.csv")
spectrograms = pd.read_csv(spectrograms)
spectrograms.head()
```

```python
# ✅ Extract & Save Clips
for spec_path in spectrograms.audiofile:
    # Parse WAV filename and clip number
    base_name = os.path.basename(spec_path)  # Extract filename
    match = re.match(r"(S4A\d+_\d{8}_\d{6})\.wav_s_(\d+)\.jpeg", base_name)
    
    if match:
        wav_filename = match.group(1) + ".wav"  # Reconstruct WAV file name
        clip_no = int(match.group(2))  # Extract clip number
        
        wav_path = os.path.join(AUDIO_DIR, wav_filename)  # Full path to WAV file
        
        if os.path.exists(wav_path):
            # ✅ Load full WAV file
            wav, sr = librosa.load(wav_path, sr=40000)  # Keep original sampling rate
            
            # ✅ Compute start and end time of the clip (1-sec clip)
            start_sample = clip_no * sr  # Start in samples
            end_sample = (clip_no + 1) * sr  # End in samples
            
            if end_sample <= len(wav):  # Ensure it does not exceed duration
                clip = wav[start_sample:end_sample]  # Extract clip
                
                # ✅ Save extracted clip
                output_filename = f"{wav_filename}_clip_{clip_no:04d}.wav"
                output_path = os.path.join(OUTPUT_DIR, output_filename)
                sf.write(output_path, clip, sr)
                
                print(f"✅ Extracted Clip: {output_path}")
            else:
                print(f"❌ Clip {clip_no} exceeds WAV duration: {wav_filename}")
        else:
            print(f"❌ WAV file not found: {wav_path}")
    else:
        print(f"❌ Filename format issue: {spec_path}")
```
